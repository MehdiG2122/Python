```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn import preprocessing
from sklearn import utils
```

#Approach with only binaries - We set negative values to 0 and positive to 1

```python
df_binary_x = pd.read_csv('/content/drive/MyDrive/dev/ML/Group/input_training.csv').fillna(0)
df_binary_y = pd.read_csv('/content/drive/MyDrive/dev/ML/Group/output_training.csv').fillna(0)
df_binary_x_test = pd.read_csv('/content/drive/MyDrive/dev/ML/Group/input_test.csv').fillna(0)
```

```python
# Keep index IDs for challenge submission
index = df_binary_x_test['ID']
```

```python
# We set in every column when return value is negative then assign 0 and 1 when positive
for colname in df_binary_x.columns:
  df_binary_x[colname] = np.where(df_binary_x[colname] > 0, 1, 0)
  df_binary_x_test[colname] = np.where(df_binary_x_test[colname] > 0, 1, 0)

df_binary_y[df_binary_y.columns[1]] = np.where(df_binary_y[df_binary_y.columns[1]] > 0, 1, 0)
```

```python
# Assigning our train and test sets
y_train = df_binary_y[df_binary_y.columns[1]]
X_train = df_binary_x[df_binary_x.columns[3:]]
X_test = df_binary_x_test[df_binary_x_test.columns[3:]]
```

```python
reg = HistGradientBoostingClassifier(loss='binary_crossentropy', learning_rate=0.05, max_iter=10000,min_samples_leaf=20, l2_regularization=0.0, max_bins=255, categorical_features=None, monotonic_cst=None, warm_start=True, early_stopping=False, scoring='loss', validation_fraction=0.1, n_iter_no_change=10, tol=1e-07, verbose=True, random_state=None)
# Train the model
reg.fit(X_train, y_train)
```

```python
# Instantiate Gradient Boosting Classifier model
reg = GradientBoostingClassifier(learning_rate=1.1,min_samples_leaf = 1, criterion='squared_error', random_state=None, max_features=None,warm_start=True, verbose=True)

# Train the model
reg.fit(X_train, y_train)
```

```python
reg.score(X_train, y_train)
```

```python
output_ypred = pd.DataFrame(reg.predict(X_test), columns=['Predicted']).set_index(index)
```
